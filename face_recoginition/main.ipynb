```python
import face_recognition
```

```python
import cv2
```

```python
known_image_array=face_recognition.load_image_file("dulq.jpg")
res=face_recognition.face_encodings(known_image_array)
if res:
    known_encoding=res[0]
```

```python
unknown_img = face_recognition.load_image_file("dulq1.jpg")
res1 = face_recognition.face_encodings(unknown_img)
if res1:
    unknown_encoding = res1[0]
```

```python
unknown_encoding=res1[0]
```

```python
face_recognition.compare_faces([known_encoding],unknown_encoding)
```

```python
y1,x2,y2,x1=face_recognition.face_locations(known_image_array)[0]
# top,right,bottom,left
cv2.rectangle(known_image_array,(x1,y1),(x2,y2),(0,0,255))
# bgr
cv2.imshow("dfd",known_image_array)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    if not ret:
        print("No frame")
        break
    h,w,_=frame.shape
    rgb_img=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    result=face_recognition.face_encodings(rgb_img)
    result1=face_recognition.face_locations(rgb_img)
    if result and result1:
        frame_encod=result[0]
        boolean=face_recognition.compare_faces([known_encoding],frame_encod)[0]
        y1,x2,y2,x1=result1[0]
        if boolean:
            cv2.putText(frame,"Dq",(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255))
            cv2.rectangle(frame,(x1,y1),(x2,y2),(255,255,255),1)
        else:
            cv2.putText(frame,"Unknown",(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255))
            cv2.rectangle(frame,(x1,y1),(x2,y2),(255,255,255),1)

    cv2.imshow('Face',frame)
    if cv2.waitKey(30)==ord("k"):
        cap.release()
        cv2.destroyAllWindows()
        break
```
