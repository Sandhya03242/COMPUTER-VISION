```python
import cv2
import re

# Define regex patterns
email_pattern = re.compile(r"[a-zA-Z0-9]+@[a-zA-Z]+\.[a-zA-Z]+")
phone_pattern = re.compile(r"\b\+?9?1?\d{10}\b")
website_pattern = re.compile(r"www\.[a-zA-Z0-9]+\.[a-zA-Z]+")

lst = []

for i in range(len(data['top'])):
    if data['conf'][i] > 20:  # Confidence threshold
        text = str(data['text'][i])  # Convert to string (in case of NaN or non-string values)
        if text and email_pattern.search(text):  # Ensure text is not empty & check for email
            x1, y1, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
            lst.append(text)
            cv2.rectangle(img_array, (x1, y1), (x1 + w, y1 + h), (0, 0, 255), 2)  # Draw bounding box

# Show image with detected emails highlighted
cv2.imshow("OCR", img_array)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Print extracted email addresses
print("Extracted Emails:", lst)
```

```python
def extract_patterns(img_array,pattern):
    var1=pytesseract.image_to_string(img_array)
    if pattern.search(var1):
        return pattern.findall(var1)
    else:
        return None
```
