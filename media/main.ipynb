```python
import mediapipe as mp
import cv2

# Initialize MediaPipe Hands
mp_hand = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
hand = mp_hand.Hands(max_num_hands=10, min_detection_confidence=0.5)

# Initialize webcam
cap = cv2.VideoCapture(0)

def left_or_right(landmarks):
    """Determine if the hand is left or right based on the x-coordinates of the wrist."""
    wrist_x = landmarks[mp_hand.HandLandmark.WRIST].x
    index_finger_x = landmarks[mp_hand.HandLandmark.INDEX_FINGER_TIP].x
    return "Left Hand" if wrist_x < index_finger_x else "Right Hand"

while True:
    ret, frame = cap.read()
    if not ret:
        print("No frame")
        break

    frame = cv2.flip(frame, 1)  # Flip the frame horizontally
    rgb_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hand.process(rgb_img)
    hand_count = 0  

    if result.multi_hand_landmarks:  
        hand_count = len(result.multi_hand_landmarks) 
        for landmarks in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(frame, landmarks, mp_hand.HAND_CONNECTIONS)
            hand_label = left_or_right(landmarks.landmark)
            cv2.putText(frame, hand_label, (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
    else:
        cv2.putText(frame, "No Hands Detected", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.putText(frame, f"Hands Detected: {hand_count}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    cv2.imshow("Hand Tracking", frame)

    if cv2.waitKey(30) == ord('k'):
        cv2.destroyAllWindows()
        cap.release()
        break
```

```python
def left_or_right(landmarks):
    thumb_tip=landmarks.landmark[4].x
    pinky_tip=landmarks.landmark[20].x
    if thumb_tip>pinky_tip:
        return "Left hand"
    else:
        return "Right hand"
```

```python

def count_fingures(landmarks):
    tips=[8,12,16,20]
    pips=[6,10,14,18]
    c=0
    for i,j in zip(tips,pips):
        if landmarks.landmark[i].y<landmarks.landmark[j].y:
            c+=1
    thumb_tip=landmarks.landmark[4].x
    pinky_tip=landmarks.landmark[20].x
    thumb_pip=landmarks.landmark[2].x
    if thumb_tip>pinky_tip:
        if thumb_tip>thumb_pip:
            c+=1
    else:
        if thumb_tip<thumb_pip:
            c+=1
    return c
```

```python
import mediapipe as mp
import cv2

# Initialize Mediapipe Hand Detector
mp_hand = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
hand = mp_hand.Hands(max_num_hands=10, min_detection_confidence=0.5)

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        print("No frame")
        break

    frame = cv2.flip(frame, 1)  # Flip horizontally
    rgb_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    result = hand.process(rgb_img)
    hand_count = 0

    if result.multi_hand_landmarks:
        hand_count = len(result.multi_hand_landmarks)
        h, w, _ = frame.shape  # Get frame height and width

        total_fingers = 0  # Count total fingers opened
        for landmarks in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(frame, landmarks, mp_hand.HAND_CONNECTIONS)

            hand_label = left_or_right(landmarks)
            x_pos = int(landmarks.landmark[0].x * w)  # Get wrist x-position
            y_pos = int(landmarks.landmark[0].y * h)  # Get wrist y-position

            # Adjust position for better readability
            if x_pos < w // 2:
                text_x = x_pos + 50  # Shift right if on left side
            else:
                text_x = x_pos - 150  # Shift left if on right side

            cv2.putText(frame, f"{hand_label}", (text_x, y_pos),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
            
            fingers_open = count_fingures(landmarks)
            total_fingers += fingers_open

        cv2.putText(frame, f"Hands Detected: {hand_count}", (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        cv2.putText(frame, f"Total Fingers Open: {total_fingers}", (20, h - 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

    else:
        cv2.putText(frame, "No Hands Detected", (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("Hand Tracking", frame)

    if cv2.waitKey(30) == ord('k'):
        cap.release()
        cv2.destroyAllWindows()
```

```python
def thumbs_up(landmarks):
    thumb_up = landmarks.landmark[4].y < landmarks.landmark[3].y
    index = landmarks.landmark[8].y > landmarks.landmark[6].y
    middle = landmarks.landmark[12].y > landmarks.landmark[10].y
    ring = landmarks.landmark[16].y > landmarks.landmark[14].y
    pinky = landmarks.landmark[20].y > landmarks.landmark[18].y

    if thumb_up and index and middle and ring and pinky:
        return True
    else:
        return False
```

```python
def thumbs_down(landmarks):
    thumb_down = landmarks.landmark[4].y > landmarks.landmark[3].y
    index = landmarks.landmark[8].y > landmarks.landmark[6].y
    middle = landmarks.landmark[12].y > landmarks.landmark[10].y
    ring = landmarks.landmark[16].y > landmarks.landmark[14].y
    pinky = landmarks.landmark[20].y > landmarks.landmark[18].y

    if thumb_down and index and middle and ring and pinky:
        return True
    else:
        return False
```

```python
def victory_sign(landmarks):
    index = landmarks.landmark[8].y < landmarks.landmark[6].y
    middle = landmarks.landmark[12].y < landmarks.landmark[10].y
    ring = landmarks.landmark[16].y > landmarks.landmark[14].y
    pinky = landmarks.landmark[20].y > landmarks.landmark[18].y

    if index and middle and ring and pinky:
        return True
    else:
        return False


```

```python
import mediapipe as mp
import cv2

# Initialize Mediapipe Hand Detector
mp_hand = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
hand = mp_hand.Hands(max_num_hands=10, min_detection_confidence=0.5)

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        print("No frame")
        break

    frame = cv2.flip(frame, 1)  # Flip horizontally
    rgb_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    result = hand.process(rgb_img)
    hand_count = 0

    if result.multi_hand_landmarks:
        hand_count = len(result.multi_hand_landmarks)
        h, w, _ = frame.shape  # Get frame height and width
        gesture = "Unknown"

        for landmarks in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(frame, landmarks, mp_hand.HAND_CONNECTIONS)

            x_pos = int(landmarks.landmark[0].x * w)  # Get wrist x-position
            y_pos = int(landmarks.landmark[0].y * h)  # Get wrist y-position

            # Detect gestures
            if thumbs_up(landmarks):
                gesture = "Thumbs Up"
            elif thumbs_down(landmarks):
                gesture = "Thumbs Down"
            elif victory_sign(landmarks):
                gesture = "Victory"

            cv2.putText(frame, f"{gesture}", (x_pos, y_pos - 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        cv2.putText(frame, f"Hands Detected: {hand_count}", (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    else:
        cv2.putText(frame, "No Hands Detected", (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("Hand Tracking", frame)

    if cv2.waitKey(30) == ord('k'):
        break

cap.release()
cv2.destroyAllWindows()
```
