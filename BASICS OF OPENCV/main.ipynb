Face Detection

```python
# pretrained model load cheyam
import cv2
import matplotlib.pyplot as plt
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face_cascade
```

```python
face_image = cv2.imread("WhatsApp Image 2025-02-08 at 14.35.23_ba3e4f28.jpg")
face_gray = cv2.cvtColor(face_image, cv2.COLOR_BGR2GRAY)

# Detect faces in the grayscale image
res = face_cascade.detectMultiScale(face_gray,minNeighbors=6)
if len(res)!=0:
    for i in res:
        x, y, w, h = i  # Unpacking 
        cv2.rectangle(face_image, (x, y), (x + w, y + h), (255, 255, 255), 2) 
cv2.imshow("face",face_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
# w=130,h=130
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    if not ret:
        print("no frame")
        break
    cv2.imshow('frame',frame)
    cv2.waitKey(100)
    if cv2.waitKey(50)==ord("k"):
        cv2.destroyAllWindows()
        cap.release()
        break
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    if not ret:
        print("No frame")
        break
    h,w,_=frame.shape
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    res=face_cascade.detectMultiScale(gray)
    if len(res)!=0:
        cv2.putText(frame,f"{len(res)} face detected",(0,h-5),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255))
        for i in res:
            x,y,w,h=i
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,255))
    else:
        cv2.putText(frame,f"no face detected",(0,h-5),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255))


    cv2.imshow('frame',frame)
    if cv2.waitKey(50)==ord('l'):
        cv2.destroyAllWindows()
        cap.release()
        break
```
